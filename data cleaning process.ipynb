```python
# tạo df_transactions + map fraud labels from json file to transactions data
import json

json_path = r"D:\Data Analyst VTI\5. Python\Topic 1. Financial Transactions Dataset Analytics\Topic 1. Financial Transactions Dataset Analytics (cut)\train_fraud_labels.json"

# Đọc dictionary từ JSON
with open(json_path, "r") as f:
    data = json.load(f)

labels_dict = data["target"]

df_transactions = pd.read_sql_query(''' 
SELECT *
FROM [topic_1].[dbo].[transactions_2012_sample_raw]
    ; ''', conn)

# Lấy label cho mỗi id
df_transactions["fraud_label"] = df_transactions["id"].astype(str).map(labels_dict)

print(df_transactions)
```

```python
df_cards[['id','client_id','cvv']] = df_cards[['id','client_id','cvv']].astype(str)
```

```python
def cvv(x):
    if len(x)==2:
        return '0'+x
    elif len(x)==1:
        return "00"+x
    elif len(x)==3 or len(x)==4:
        return x
    
    
df_cards['cvv']=df_cards['cvv'].apply(lambda x : cvv(x))
```

# 4. STORE CLEAR DATA INTO SQL SERVER

```python
# --- Connect ---
conn = pyodbc.connect(
    "DRIVER={ODBC Driver 17 for SQL Server};"
    "SERVER=localhost\SQLEXPRESS;"
    "DATABASE=topic_1;"
    "Trusted_Connection=yes;"
    "TrustServerCertificate=yes;"
)
cursor = conn.cursor()

# --- Map pandas dtype -> SQL Server type ---
dtype_map = lambda dt: "BIGINT" if pd.api.types.is_integer_dtype(dt) else \
                       "DECIMAL(18,4)" if pd.api.types.is_float_dtype(dt) else \
                       "BIT" if pd.api.types.is_bool_dtype(dt) else \
                       "DATETIME" if pd.api.types.is_datetime64_any_dtype(dt) else \
                       "VARCHAR(255)"

table_name = "cards_data_clear"

# --- Drop + Create table ---
cursor.execute(f"IF OBJECT_ID('dbo.{table_name}', 'U') IS NOT NULL DROP TABLE dbo.{table_name}")
cols = ", ".join([f"[{c}] {dtype_map(df_cards[c].dtype)}" for c in df_cards.columns])
cursor.execute(f"CREATE TABLE dbo.{table_name} ({cols})")

# --- Insert rows ---
cursor.fast_executemany = True
cursor.executemany(f"INSERT INTO dbo.{table_name} VALUES ({','.join(['?']*len(df_cards.columns))})", df_cards.values.tolist())
conn.commit()
```
