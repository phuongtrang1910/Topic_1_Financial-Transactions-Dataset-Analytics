```python
import pandas as pd

input_file = r"D:\Data Analyst VTI\5. Python\Topic 1. Financial Transactions Dataset Analytics\Topic 1. Financial Transactions Dataset Analytics (cut)\transactions_data.csv"
output_file = r"D:\Data Analyst VTI\5. Python\Topic 1. Financial Transactions Dataset Analytics\Topic 1. Financial Transactions Dataset Analytics (cut)\transactions_2012_sample.csv"

chunksize = 500_000
date_col = "date"
max_rows = 900_000
written_rows = 0  # đếm số dòng đã ghi

for chunk in pd.read_csv(input_file, chunksize=chunksize, parse_dates=[date_col]):
    # Lọc theo năm 2012
    filtered = chunk[(chunk[date_col] >= "2012-01-01") & (chunk[date_col] <= "2012-12-31")]

    if filtered.empty:
        continue

    # Nếu ghi hết filtered vượt quá max_rows, chỉ lấy phần cần thiết
    if written_rows + len(filtered) > max_rows:
        filtered = filtered.iloc[:max_rows - written_rows]

    # Ghi vào CSV
    filtered.to_csv(output_file, mode="a", index=False, header=(written_rows==0))

    # Cập nhật số dòng đã ghi
    written_rows += len(filtered)

    # Dừng khi đã đủ 900,000 dòng
    if written_rows >= max_rows:
        break

print(f"Finished writing {written_rows} rows to {output_file}")
```
